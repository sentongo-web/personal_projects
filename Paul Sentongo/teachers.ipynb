```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

Reading the csv file

```python
teacher_ratings = pd.read_csv('teachingratings.csv')
```

visualizing the content of the dataset

```python
teacher_ratings
```

##### Question 1: Create a distribution plot of beauty scores with Native English speaker as a factor

```python
# Setting the style
sns.set(style="whitegrid")

# Seting the figure size
plt.figure(figsize=(10, 6))

# Getting unique values in the 'English_speaker' column
english_speaker_values = teacher_ratings['English_speaker'].unique()

# Creating a palette with different colors for each unique value
palette = sns.color_palette("husl", len(english_speaker_values))

# Creating a distribution plot with the dynamic palette
sns.histplot(data=teacher_ratings, x='beauty', hue='English_speaker', kde=True, bins=20, stat="density", common_norm=False, palette=dict(zip(english_speaker_values, palette)))

# Setting major title
plt.suptitle("A Histogram with Kernel Density Estimates for Beauty Scores", fontsize=16, fontweight='bold')

# Setting subtitle
plt.title("Distribution of Beauty Scores by Native English Speaker Status", fontsize=14)

# Setting plot labels
plt.xlabel('Beauty Score')
plt.ylabel('Density')

# Showing the plot
plt.show()
```

##### Question 2: Create a Horizontal box plot of the age of the instructors by visible minority

```python
# Setting the style
sns.set(style="whitegrid")

# Setting the figure size
plt.figure(figsize=(10, 6))

# Getting unique values in the 'minority' column
minority_values = teacher_ratings['minority'].unique()

# Creating a palette with different colors for each unique value
palette = sns.color_palette("husl", len(minority_values))

# Creating a horizontal box plot with the dynamic palette
sns.boxplot(data=teacher_ratings, x='age', y='minority', orient='h', palette=dict(zip(minority_values, palette)))

# Setting plot labels and title
plt.xlabel('Age')
plt.ylabel('Visible Minority')
plt.title('Horizontal Box Plot of Age by Visible Minority Status')

# Showing the plot
plt.show()
```

##### Question 3: Create a group histogram of tenure by minority and add the gender factor

```python
# Setting the style
sns.set(style="whitegrid")

# Setting the figure size
plt.figure(figsize=(12, 8))

# Creating a grouped histogram with different colors
g = sns.histplot(data=teacher_ratings, x='tenure', hue='minority', multiple='stack', binwidth=1, stat="count", common_norm=False, palette="Set2")

# Adding gender factor as different shades within each bar
sns.histplot(data=teacher_ratings, x='tenure', hue='gender', multiple='stack', binwidth=1, stat="count", common_norm=False, palette="Set1", alpha=0.7, ax=g)

# Setting plot labels and title
plt.xlabel('Tenure')
plt.ylabel('Count')
plt.title('Grouped Histogram of Tenure by Minority with Gender Factor')

# Adding legend
plt.legend(title='Gender', loc='upper right', labels=['Male', 'Female'])

# Showing the plot
plt.show()
```

##### Question 4: Create a boxplot of the age variable

```python
# Setting the style
sns.set(style="whitegrid")

# Setting the figure size
plt.figure(figsize=(8, 6))

# Creating a boxplot for the 'age' variable with red color
sns.boxplot(x=teacher_ratings['age'], color='red')

# Setting plot labels and title
plt.xlabel('Age')
plt.title('Boxplot of Age')

# Showing the plot
plt.show()
```
