##### Below i create a list that contains all names of my datasets and then i store them,
##### I load the datasets into the list and then i add a newcolumn, which column is to have values of cumulative population and Save the result in a CSV file called fao.csv.


```python
# List of dataset filenames
dataset_files = ["Nigeria.csv", "India.csv", "Uk.csv", "USA.csv", "Portugal.csv"]

# List to store loaded datasets
datasets = []

# Load each dataset into the list and add 'newColumn', we are to use a loop to check if the field Value 
# exists in the dataset and then it loads data
for file in dataset_files:
    # just assuming if all datasets have a header row
    fao = pd.read_csv(file)
    
    # Checking if 'Value' column exists before adding 'newColumn' then storing values of cummulative population there
    if 'Value' in fao.columns:
        fao['newColumn'] = fao['Value'].cumsum()
    else:
        print(f"Warning: 'Value' column not found in {file} dataset.")
    
    datasets.append(fao)

# Concatenating all dataframes into a single dataframe named fao
fao = pd.concat(datasets, ignore_index=True)

# Saving the concatenated dataframe to a CSV file named 'fao.csv'
fao.to_csv("fao.csv", index=False)

# Displaying the head of the modified dataframe fao
print("\nHead of the 'fao' dataset with 'newColumn' (Cumulative Population):\n")
print(fao.head().to_string(index=False))
```

##### 2.	With fao, use type conversion to cast the values of the year column into datetimes and, sort by date and value.

##### A new sorted dataframe is later created

```python
# List of dataset filenames
dataset_files = ["Nigeria.csv", "India.csv", "Uk.csv", "USA.csv", "Portugal.csv"]

# List to store loaded datasets
datasets = []

# Load each dataset into the list and add 'newColumn'
for file in dataset_files:
    # Assuming all datasets have a header row
    fao = pd.read_csv(file)
    
    # Checking if 'Value' column exists before adding 'newColumn'
    if 'Value' in fao.columns:
        fao['newColumn'] = fao['Value'].cumsum()
    else:
        print(f"Warning: 'Value' column not found in {file} dataset.")
    
    datasets.append(fao)

# Concatenating all dataframes into a single dataframe
fao = pd.concat(datasets, ignore_index=True)

# Converting the 'Year' column to datetime format
fao['Year'] = pd.to_datetime(fao['Year'], format='%Y')

# Sorting the dataframe by 'Year' and 'Value'
fao = fao.sort_values(by=['Year', 'Value'])

# Saving the sorted dataframe to a CSV file named 'fao_sorted.csv'
fao.to_csv("fao_sorted.csv", index=False)

# Displaying the head of the sorted dataframe
print("\nHead of the sorted 'fao' dataset by 'Year' and 'Value':\n")
print(fao.head().to_string(index=False))
```

### DATA WRANGLING TECHNIQUES

##### Technique A: Handling Missing Values
Checking for missing values in the dataset and decide on an appropriate strategy for handling them, and imputing has been done

```python
# Checking for missing values in the dataset
missing_values = fao.isnull().sum()

# Printing the number of missing values for each column
print("\nNumber of missing values for each column:")
print(missing_values)

# Imputing missing values in 'Value' column with mean
fao['Value'].fillna(fao['Value'].mean(), inplace=True)

# Displaying the modified dataframe after handling missing values
print("\nHead of the modified 'fao' dataset after handling missing values:")
print(fao.head())
```

##### Technique B: Handling Outliers
Identifying and handling outliers in the data. Outliers can significantly impact statistical measures and machine learning models.

```python
import pandas as pd
import numpy as np

# Loading the 'fao' dataset
fao = pd.read_csv("fao_sorted.csv")

# Identifying and handle outliers using winsorization
def winsorize(series, lower_limit=0.05, upper_limit=0.95):
    lower_bound = series.quantile(lower_limit)
    upper_bound = series.quantile(upper_limit)
    return np.clip(series, lower_bound, upper_bound)

# Applying winsorization to the 'Value' column
fao['Value'] = winsorize(fao['Value'])

# Displaying summary statistics after handling outliers
print("\nSummary Statistics after handling outliers:")
print(fao.describe())
```

##### Technique C: Data Normalization or Standardization
This is particularly important for my dataset which contains numerical columns with different scales. 

```python
from sklearn.preprocessing import MinMaxScaler
# Normalize 'Value' column using Min-Max scaling
scaler = MinMaxScaler()
fao['Value_normalized'] = scaler.fit_transform(fao[['Value']])

# Display the modified dataframe with normalized values
print("\nHead of the modified 'fao' dataset with normalized values:")
print(fao.head())
```

##### Technique D: Handling Duplicate Data
Checking for any and handling any duplicate entries in the dataset. This is because Duplicates can skew analysis and introduce biases. That is why i Remove or consolidate duplicate rows as appropriate.

```python
# Check for and handle duplicate entries
duplicates_count = fao.duplicated().sum()

# Print the number of duplicate entries
print("\nNumber of duplicate entries:", duplicates_count)

# Drop duplicate rows
fao_no_duplicates = fao.drop_duplicates()

# Display the modified dataframe without duplicate entries
print("\nHead of the modified 'fao' dataset without duplicates:")
print(fao_no_duplicates.head())
```

##### Pivoting Data (Technique - E):

Pivoting the data can be useful for changing its structure. Depending on my analysis needs, i want to pivot the data to a wider or longer format.

```python
# To pivot on 'Element' and 'Year'
fao_pivoted = fao.pivot(index='Area', columns='Year', values='Value')

# DisplayING the pivoted dataframe
print("\nHead of the pivoted 'fao' dataset:")
print(fao_pivoted.head())
```

##### Aggregating Data (Technique - F):
This involves grouping by certain columns and calculating summary statistics

```python
# I want to calculate the mean 'Value' for each 'Area'
fao_aggregated = fao.groupby('Area')['Value'].mean().reset_index()
# Displaying the aggregated dataframe
print("\nHead of the aggregated 'fao' dataset:")
print(fao_aggregated.head())
```

Technique G: Handling Skewed Data

```python
# Applying log transformation to 'Value' column
fao['Value_log_transformed'] = np.log1p(fao['Value'])

# Displaying the modified dataframe with log-transformed values
print("\nHead of the modified 'fao' dataset with log-transformed values:")
print(fao.head())
```
